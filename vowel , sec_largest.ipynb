```python
def find_vowel(string:str) -> str:
    if not isinstance(string,str):
        raise TypeError("Input must be a string")
    if len(string) == 0:
        raise ValueError("input is empty")
    vowel = "aeiou"
    count = 0
    for char in string.lower():
        if char in vowel:
            count = count + 1
    return count
# ___ main__ excuetion   
if __name__ == "__main__":
    try: 
        user_input = input("enter your string")
        result = find_vowel(user_input)
        print(f"Vower is : ,{result}")
        
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
```

```python
def second_largest(number_list :list) -> int:
    if not isinstance(number_list,list):
        raise TypeError("input must be a list")
    if len(number_list) <2 :
        raise ValueError("Input list must contain atleat 2 arguments of number")
    if not all(isinstance(current_number,(int,float)) for current_number in number_list):
        raise TypeError("All elements must be a integer")
    if number_list[0] > number_list[1]:
        largest = number_list[0]
        second = number_list[1]
    else:
        largest = number_list[1]
        second = number_list[0]
    for current_number in number_list:
        if current_number > largest:
            second = largest
            largest = current_number
        elif current_number > second and current_number != largest:
            second = current_number
    return second
if __name__ == "__main__":
    number_list = [1,2,4,56,67,87]
    try:
        result = second_largest(number_list)
        print(f"second largest number is ,{result}")
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
    
    
    
```
